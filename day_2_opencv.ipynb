```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# function for displaying an image
def imshow(title="", image=None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
#Drawing shapes and forms 
image_with_line = np.zeros((512, 512, 3), np.uint8)
imshow("Black image", image=image_with_line)
```

```python
cv2.line(image_with_line, (0,0),(511,511),(255,127,0),5)
imshow("Black Canvas with Diagonal Line", image=image_with_line)
```

```python
# Mettre un rectangle dans une image 
image_with_rectangle = np.zeros((512,512,3),np.uint8)
cv2.rectangle(image_with_rectangle, (100, 100),(400,250),(127,50,127),10)
imshow("Black image with pink rectangle", image_with_rectangle)
```

```python
# Mettre un cercle dans une image 
image_with_circle = np.zeros((512,512,3),np.uint8)
cv2.circle(image_with_circle, (200, 300),100,(15,150,50),-1)
imshow("Black image with pink rectangle", image_with_circle)
```

```python
# Ecrire dans une image
image_with_text = np.zeros((600,512,3),np.uint8)
outString = "Hello world"
cv2.putText(image_with_text,outString ,(30,300), cv2.FONT_HERSHEY_SCRIPT_COMPLEX,3,(40,200,0),4)
imshow("Black image with pink rectangle", image_with_text)
```

```python
# Tourner une image
image = cv2.imread("Photos/aircraft-sales-3.jpg")
height, width = image.shape[:2]

rotation_matrix = cv2.getRotationMatrix2D(center=(width/2, height/2),angle=90,scale=1)
```

```python
rotated_image = cv2.warpAffine(image, rotation_matrix, (width,height))
imshow("Rotated 90 with scale 1", rotated_image)
```

```python
#Transposition 
rotated_image_2 = cv2.transpose(image)
imshow("original image", image=image)
imshow("rotated image", rotated_image_2)
```

```python
# diminuer la taille d'une image 
image_scaled = cv2.resize(image, None, fx=.75, fy=.75 )
imshow("0.70X Scaling - Linear interpolation", image_scaled)
```

```python
# diminuer la taille d'une image 
image_scaled = cv2.resize(image, None, fx=2, fy=2 )
imshow("2X Scaling - Linear interpolation", image_scaled)
```

```python
# diminuer la taille d'une image 
image_scaled = cv2.resize(image,(900, 400),interpolation=cv2.INTER_AREA)
imshow("0.70X Scaling - Linear interpolation", image_scaled)
```

```python
# PYRAMINDS
image = cv2.imread("Photos/mpicture.jpg")

smaller = cv2.pyrDown(image)
larger = cv2.pyrUp(smaller)

imshow("Original", image=image)
imshow(title="smaller", image=smaller)
imshow(title="larger", image=larger)

even_smaller = cv2.pyrDown(smaller)
imshow(title="even smaller", image=even_smaller)
```

CROPPING

```python
image = cv2.imread('Photos/apple.jpeg')
height, width = image.shape[:2]

start_row, start_col = int(height*.25), int(width*.25)
end_row, end_col = int(height*.75), int(width*.75)

cropped = image[start_row:end_row, start_col:end_col]

imshow("Original image", image=image)

copy = image.copy()
cv2.rectangle(img=copy,pt1=(start_col,start_row),pt2=(end_col,end_row),color=(0,255,255),thickness=10)
imshow("Area to be cropped", copy)
imshow("Cropped image", cropped)
```

```python
# Ajouter, retrancher de la luminosite 
image = cv2.imread("Photos/pineapple.jpeg")
imshow("Gray scaled", image=image)
```

```python
# Creer une matrice de 1 avec les memes dimensions que notre photo
M = np.ones(image.shape, dtype="uint8")*100

# Augmenter la luminosite 
added =cv2.add(image, M)
imshow("Increased brightness", added)

added2 = image+M
imshow("Simple numpy added brightness", added)
```

```python
substracted = cv2.subtract(image, M)

imshow("Substracted", substracted)

substracted = image-M
imshow("Substracted", substracted)
```
